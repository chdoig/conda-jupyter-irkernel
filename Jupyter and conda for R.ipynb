# Jupyter and conda for R

This notebook explains the many benefits Jupyter, the IRKernel and conda can provide for data scientists working in R.

[Jupyter](http://jupyter.org/), previously called IPython, is already widely adopted by data scientists, researchers, and analysts. Jupyter's notebook user interface enables mixing executable code with narrative text, equations, interactive visualizations, and images to enhance team collaboration and advance the state of reproducible research and training. Jupyter began with Python and now has kernels for 50 different languages, and the [IRKernel](http://irkernel.github.io/) is the native R kernel for Jupyter.

Data scientists, researchers, and analysts use the [conda](http://conda.io/) package manager to install and organize project dependencies. With conda they can easily build and share metapackages, which are downloadable bundles of packages. Conda works with Linux, OS X, and Windows, and is language agnostic, so we can use it with any programming language and with projects that depend on multiple languages.


Let's use conda and Jupyter to start a data science project in R.

## "R Essentials" setup

The Anaconda team has created an ["R Essentials"](http://anaconda.org/r/r-essentials) bundle with the IRKernel and over 80 of the most used R packages for data science, including `dplyr`, `shiny`, `ggplot2`, `tidyr`, `caret` and `nnet`.


Downloading "R Essentials" requires conda. [Miniconda](http://conda.pydata.org/miniconda.html) includes conda, Python, and a few other necessary packages, while [Anaconda](https://store.continuum.io/cshop/anaconda/) includes all this and over 200 of the most popular [Python packages](http://docs.continuum.io/anaconda/pkg-docs.html) for science, math, engineering, and data analysis. Users may install all of Anaconda at once, or they may install Miniconda at first and then use conda to install any other packages they need, including any of the packages in Anaconda.

Once you have conda, you may install "R Essentials" into the current environment:


```bash
conda install -c r r-essentials 
```


or create a new environment just for "R essentials":

```bash
conda create -n my-r-env -c r r-essentials 
```

## An R notebook example

Now you can:

- import the data wrangling R package, dplyr:

````r
library(dplyr)
````

- explore one of the available datasets, such as the `iris`:

````r
iris
````

- calculate the average sepal width by species:


````r
iris %>%
 group_by(Species) %>%
 summarise(Sepal.Width.Avg = mean(Sepal.Width)) %>%
 arrange(Sepal.Width.Avg)
````

- import the R visualization library ggplot:

````r
library(ggplot2)
````

- plot the Sepal.Width vs. Sepal.Length

````r
ggplot(data=iris, aes(x=Sepal.Length, y=Sepal.Width, color=Species)) + geom_point(size=3)
````

## Creating custom R bundles


For our users' convenience, we have packaged some of the most used R packages for data science in "R Essentials". It is also very easy to create your own custom set of R packages to share with peers with the `conda metapackage` command. For example, to provide a download called custom-r-bundle with only the libraries used in our example notebook, just create the metapackage:

```bash
conda metapackage custom-r-bundle 0.1.0 --dependencies r-irkernel jupyter r-ggplot2 r-dplyr --summary "My custom R bundle"
```


Share it with colleagues by uploading it to [Anaconda.org](https://anaconda.org/):

```bash
conda install anaconda-client
anaconda login
anaconda upload custom-r-bundle-0.1.0-0.tar.bz2
```

Now anyone can get the bundle with those packages and dependencies by replacing "anacondauser" with your Anaconda.org username and running this command:

```bash
conda install -c anacondauser custom-r-bundle
```
